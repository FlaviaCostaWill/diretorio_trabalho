```python
!pip install awswrangler

import awswrangler as wr
```

```python
def query(query, username):
    """Retorna df da consulta ao banco de dados"""
    
    from datetime import datetime
    dt = datetime.today().strftime('%Y-%m-%d-%H-%M-%S-%f')
    path = f"s3://will-sdx-ml-platform-sagemaker-transient/{username}/{dt}"

    df = wr.athena.read_sql_query(query, 
    database=None, 
    ctas_approach=False, 
    s3_output=path)

    return df
```

```python
dias_base = 10
```

```python
query_string = f"""
    
with base_chats as (
   select distinct 
cast(issue.id_issue as varchar) as id_issue
	, dt_envio_mensagem
	, date_trunc('month', issue.dt_criacao_chat) as mes_chat
	,  ds_message
    , is_resolvido_bot
	from cx_curated_zone.helpshift_issues_messages msg
	left join cx_curated_zone.helpshift_issues issue on (issue.id_issue  = msg.id_issue)
	where dt_envio_mensagem >= date_trunc('month', date_add('month', (-1*{dias_base}),now()))
	and ds_entidade = 'agente'
    and issue.is_cx_ou_cobranca = 'CX'
union all
---------------INTERCOM -------------------
select distinct 
m.id_issue 
	, m.dt_envio_mensagem
	, date_trunc('month', cast(i.dt_criacao as timestamp)) as mes_chat
	,  ds_message
    , is_resolvido_bot
	from cx_curated_zone.intercom_issues_messages m
left join cx_curated_zone.intercom_issues_id i on (i.id_issue = m.id_issue)
	where dt_envio_mensagem >= date_trunc('month', date_add('month', (-1*{dias_base}),now()))
	and m.ds_entidade = 'agente'
    and i.is_cx_ou_cobranca = 'CX'
    )
   select * from (
    select 
        msg.id_issue
        , dt_envio_mensagem
        , ds_message
        , msg.is_resolvido_bot 
       -- Tags
       , i.nm_tag_hierarquia_4 as hierarquia_4
       , i.nm_tag_hierarquia_3 as hierarquia_3
       , i.nm_tag_hierarquia_2 as hierarquia_2
       , i.nm_tag_hierarquia_1 as hierarquia_1
    from 
       base_chats msg
   left join cx_curated_zone.tags_helpshift_intercom i on i.id_issue = msg.id_issue
        )
    order by rand() desc
    limit 100

    """

df = query(query_string, 'flavia-costa')
```
