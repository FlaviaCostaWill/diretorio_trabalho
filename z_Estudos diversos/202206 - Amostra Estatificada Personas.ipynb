```python
import pandas as pd
import numpy as np
import aux.acessos as ac
import datetime
```

```python
query = "select * from growth_curated_zone.proposal_general where dt_approval is not null limit 10000"

base = ac.df_athena('flavia-costa', query)
```

```python
base['safra_ano'] = base['dt_approval'].astype("string").str[:4].astype("int")
base['safra'] = base['dt_approval'].astype("string").str[:7].str.replace('-','').astype("int")
```

```python
#amostra da base inteira
amostragem = 0.2

#definindo pesos por safra
def weight_match(safra):
    if   safra >= 201901 and safra < 202001: #ano de 2019
        matchVar = 0.1
    elif safra >= 202001 and safra < 202101: #ano de 2020
        matchVar = 0.2
    elif safra >= 202101 and safra < 202106: #ano de 2021-S1
        matchVar = 0.1
    elif safra >= 202106: #ano de 2021-S2
        matchVar = 0.6
    else:
        matchVar = 0
    return matchVar
```

```python
base['weight'] = base.apply(lambda x: weight_match(x['safra']), axis=1) 
base['weight_final'] = base['weight'] * amostragem
base['qtde_max'] = base['weight_final'] * len(base)
base['aleat'] = np.random.random((len(base),1))
base['rank'] = base['aleat'].rank()
base_amostra_1 = base.query('rank <= qtde_max')
```
