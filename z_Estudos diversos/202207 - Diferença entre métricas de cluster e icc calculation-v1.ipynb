```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
df_model = df.query('n_cluster == "gowers_15"')
df_model = df_model.drop(['n_cluster'], axis = 1).set_index('var_value').transpose().reset_index()
preditores = df_model.drop(['index'], axis = 1).columns
x_train, x_test, y_train, y_test = train_test_split(df_model[preditores]
                                                    , df_model['index']
                                                    , random_state=42, test_size=1)
scaler = StandardScaler()
scaler_model = scaler.fit(x_train)
x_train_scaled = scaler_model.transform(x_test)
y_train = y_test
arvores = [20, 50, 100, 300, 400, 500]
accuracy_array = []

for i in arvores:
    rf = RandomForestClassifier(n_estimators=i)
    rf.fit(x_train_scaled, y_train)
    y_predicted_rf = rf.predict(x_train_scaled)
    accuracy_rf = metrics.accuracy_score(y_train, y_predicted_rf).round(4)
    accuracy_array.append(accuracy_rf)

rf_estimators = pd.DataFrame(list(zip(arvores, accuracy_array)), columns =['arvores', 'acuracia'])
rf_estimators.head(10)
```
