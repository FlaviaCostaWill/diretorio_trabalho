```python
import acessos as ac
```

```python
query  = '''

    select 
    base
    , segmento_final
    , score
    , count(distinct c.cpf) as clientes
    from customer_sandbox_zone.mgm_pesquisa p
    left join growth_curated_zone.clientes c on (c.email = p.e_mail)
    left join customer_sandbox_zone.segmentacao_mgm m on (m.cpf = c.cpf)
    group by 1,2,3

'''
```

```python
dfagg = ac.df_athena_q('flavia-costa', query)
```

```python
dfagg['clientes'].sum()
```

```python
pd.crosstab(index = dfagg.segmento_final, columns= dfagg.base , values = dfagg.clientes, aggfunc=np.sum)
```
